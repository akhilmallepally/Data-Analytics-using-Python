Objective : Given the features(age, year and nodes), classify it's status(1 or 2).

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

hab = pd.read_csv("./haberman.csv")
print(hab.head())
```

```python
print(f'(rows, columns) : {hab.shape}') #rows represent no. of data points or observations ans columns represent features
print(f'column names in the dataset : {hab.columns}')
```

```python
#(Q) No of datapoints for each class or label
#Here status is the class or label
print(hab["status"].value_counts())
```

```python
hab.plot(kind='scatter', x = 'age', y = 'nodes')
plt.show()
#Data points are 
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(hab, hue='status', height = 3) \
    .map(plt.scatter, "age", "nodes") \
    .add_legend()
plt.show()
```

```python
#Pair Plot
plt.close()
sns.set_style("whitegrid")
sns.pairplot(hab, hue = "status", height = 3)
plt.show()
```
